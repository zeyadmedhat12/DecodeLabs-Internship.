```python
d.isna().sum()
# there are 309 missing values in CouponCode
```

```python
d["Actual Price"] = d["Actual Price"].fillna(d["Actual Price"].mean())
d["Rating"] = d["Rating"].fillna(d["Rating"].mean())
d["Reviews"] = d["Reviews"].fillna(d["Reviews"].mean())
d.drop(columns=["Unnamed: 0","Name", "Model"], inplace=True)
d["Core"] = d["Core"].fillna("Unknown")
d["SSD"] = d["SSD"].fillna("Unknown")
d["Saving"] = d["Saving"].str.replace("% OFF", "").str.strip().astype(float)
d["Saving"] = d["Saving"].fillna(d["Saving"].mean())
```

```python
d.isna().sum()
```

```python
d.duplicated().sum()
# 3 duplicates found
```

```python
d.drop_duplicates(inplace=True)
d.duplicated().sum()
# all duplicates removed
```

```python

d["Brand"] = d["Brand"].str.strip().str.title()
d["Core"] = d["Core"].str.strip().str.title()
d["SSD"] = d["SSD"].str.strip()

d["Discounted Price"] = pd.to_numeric(d["Discounted Price"], errors="coerce")
d["Actual Price"] = pd.to_numeric(d["Actual Price"], errors="coerce")

d.reset_index(drop=True, inplace=True)
d.head()
```
